```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np

# 하트 모양의 마스크 이미지 파일 경로
heart_mask_path = 'heart_mask.png'  # 하트 모양 이미지 파일

# 마스크 이미지 로드 및 변환
heart_mask = np.array(Image.open(heart_mask_path))

# 텍스트 데이터
text = """ Python is a great programming language. Python is popular for data analysis,
machine learning, and web development. Python is easy to learn and has powerful libraries.
"""

# 워드 클라우드 생성
wordcloud = WordCloud(
    width=800,        # 가로 크기
    height=800,       # 세로 크기
    background_color='white',  # 배경 색상
    mask=heart_mask,  # 마스크 모양 적용
    contour_width=1,  # 테두리 선 굵기
    contour_color='white',  # 테두리 색상
    font_path='NanumGothic'  # 한글 폰트 설정 (예: 'malgun.ttf' 또는 시스템에 설치된 한글 폰트)
).generate(text)

# 워드 클라우드 시각화
plt.figure(figsize=(10, 10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')  # 축 제거
plt.show()
```

```python
!pip install networkx
import networkx as nx
G = nx.DiGraph()
G.add_nodes_from(['A','B','C','D','E'])
G.add_edges_from([('A','B'), ('B','C'), ('C','D')])
nx.draw(G, with_labels=True)
```

```python
#노드 추가 및 제거하기
g1 = nx.Graph()
g1.add_node('David')
g1.add_node('John')
g1.add_node('Mary')
g1.add_node('Julia')
g1.add_node('Juda')
g1.add_nodes_from(['Judy', 'Karen'])
g1.remove_node('Julia')
#선 추가 및 제거하기
g1.add_edge('David', 'John')
g1.add_edge('David', 'Mary')
g1.add_edge('David', 'Karen')
g1.add_edges_from([('John', 'Mary'), ('John', 'Judy')])
g1.remove_edge('Mary', 'John')
nx.draw(g1, with_labels=True, font_weight='bold')
```

```python
d = dict(g1.degree)
nx.draw(g1, nodelist=d.keys(), node_size=[v * 1000 for v in d.values()],\
with_labels=True, font_weight='bold')
g1.nodes
```

```python
g2 = nx.DiGraph()
g2.add_edges_from([(1, 2), (1, 3), (1, 4), (3, 4)])
nx.draw(g2, with_labels=True, font_weight='bold')
print(g2.in_degree)
print(g2.out_degree)
```

```python
g3 = nx.DiGraph()
g3.add_weighted_edges_from([(1, 2, 10), (2, 3, 20)])
g3.add_edge(1, 3, weight=30)
pos = nx.spring_layout(g3)
nx.draw(g3, pos=pos, with_labels=True)
labels = nx.get_edge_attributes(g3, 'weight')
nx.draw_networkx_edge_labels(g3, pos, edge_labels=labels);
```
