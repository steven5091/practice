```python
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# 데이터 정의
x = [[2], [4], [6], [8], [10]]  # 독립변수: 공부 시간
y = [[81], [93], [90], [97], [100]]  # 종속변수: 시험 점수

# 산점도 그래프
plt.scatter(x, y)
plt.xlabel('Study Hours')  # X축 라벨
plt.ylabel('Test Scores')  # Y축 라벨
plt.show()

# 학습시키기
model = LinearRegression()  # 선형 회귀 분석 객체 생성
model.fit(x, y)  # 학습하기

# 예측하기
result = model.predict([[7]])  # 7시간 공부했을 때 점수 예측
print(f'예상 점수: {result[0][0]}')  # 예측 결과 출력
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import seaborn as sns
seoul= pd.read_excel('/content/(2010-2020) weather.xlsx')
seoul.head()
seoul.info()
seoul.describe()
```

```python
seoul.drop('지점',axis=1,inplace=True)
seoul.head()
```

```python
seoul.colunms=['날짜','평균기온','최저기온','최고기온']
seoul.colunms
```

```python
seoul.isnull().sum()
```

```python
seoul.dropna(subset=['최고기온(℃)'],axis=0,inplace=True)
seoul.isnull().sum()
```

```python
seoul['년도']=seoul['날짜'].dt.year
seoul.head()
```

```python
conditions=(seoul['날짜'].dt.month==8)&(seoul['날짜'].dt.day==15)
conditions
seoul10815=seoul[conditions]
seoul10815
```

```python
fig=plt.figure(figsize=(15,7))
X=seoul10815[['년도']]
Y=seoul10815[['평균기온(℃)']]
plt.scatter(X,Y)
plt.xlabel('년도')
plt.ylabel('평균기온')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
X=seoul10815[['년도']]
Y=seoul10815[['평균기온(℃)']]
model.fit(X,Y)
result = model.predict([[2022]])  # Changed '2022' to 2022
print(result)
```

```python
x=seoul['년도'].values # Extract the values as a 1D NumPy array
y=seoul['평균기온(℃)'].values # Extract the values as a 1D NumPy array
fp1=np.polyfit(x,y,2)
f1=np.poly1d(fp1) # Corrected typo: polyId to poly1d
fx=np.linspace(2010,2020)
plt.figure(figsize=(15,7))
plt.scatter(x,y)
plt.plot(fx,f1(fx),ls='dashed',lw=3,color='g')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(15,7))

X=seoul[['년도','최저기온(℃)','최고기온(℃)']]
Y=seoul[['평균기온(℃)']]

x_train,x_test,y_train,y_test=train_test_split(X,Y,train_size=0.7,test_size=0.3)
model=LinearRegression()
model.fit(x_train,y_train)

plt.plot(model.predict(x_test[:50]),label='predict')
plt.plot(y_test[:50].values.reshape(-1,1),label='real temp')
plt.legend()
```

```python
plt.figure(figsize=(15,7))
y_predict=model.predict(x_test)

plt.scatter(y_test,y_predict,alpha=0.4)
plt.show()
```

```python
print(model.score(x_train,y_train))
```
